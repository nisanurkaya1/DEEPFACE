```python
import tensorflow as tf
import os
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

BATCH = 16
IMG_SIZE = (224, 224)
AUTOTUNE = tf.data.AUTOTUNE

DATASET_PATH = '/content/drive/MyDrive/FaceForensics_Data/Hoca_Image_Dengeli_V7'

# Yol kontrolü
for subdir in ["train", "val"]:
    path = os.path.join(DATASET_PATH, subdir)
    if not os.path.exists(path):
        raise FileNotFoundError(f"Bulunamadı: {path}")

train_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "train"),
    label_mode="binary",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=True
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "val"),
    label_mode="binary",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=False
)

train_ds = train_ds.cache().shuffle(3000).prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)

# Veri artırma azaltıldı
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),  # Azaltıldı
    layers.RandomZoom(0.1),      # Azaltıldı
    layers.RandomContrast(0.2),
    layers.RandomBrightness(0.2)
])

base = EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3)
)

base.trainable = True
for layer in base.layers[:-150]:
    layer.trainable = False

inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = layers.Lambda(preprocess_input)(x)
x = base(x, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.7)(x)  # Artırıldı
outputs = layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)

loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.05)  # Azaltıldı

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),  # Artırıldı
    loss=loss_fn,
    metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.3,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/FaceForensics_Data/best_model_full_frame_v2.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

print("🚀 DÜZELTİLMİŞ MODEL BAŞLIYOR")

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=200,
    callbacks=callbacks
)
```

```python
import tensorflow as tf
import os
import numpy as np
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

BATCH = 16
IMG_SIZE = (224, 224)
AUTOTUNE = tf.data.AUTOTUNE

DATASET_PATH = '/content/drive/MyDrive/FaceForensics_Data/Hoca_Image_Dengeli_V7'

train_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "train"),
    label_mode="binary",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=True,
    seed=123
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "val"),
    label_mode="binary",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=False
)

# Class weights manuel hesapla
train_labels = np.concatenate([y.numpy() for x, y in train_ds], axis=0)
n_samples = len(train_labels)
n_class_0 = np.sum(train_labels == 0)
n_class_1 = np.sum(train_labels == 1)
class_weight_dict = {0: n_samples / (2 * n_class_0), 1: n_samples / (2 * n_class_1)}

train_ds = train_ds.cache().shuffle(3000).prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.05),
    layers.RandomContrast(0.1),
])

base = EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3)
)
base.trainable = False

inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = layers.Lambda(preprocess_input)(x)
x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.3,
        patience=4,
        min_lr=1e-7,
        verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/FaceForensics_Data/best_model_full_frame_v3.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

print("🚀 DENGELİ MODEL BAŞLIYOR")

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    class_weight=class_weight_dict,
    callbacks=callbacks
)
```

```python
import tensorflow as tf
import os
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

# =========================
# PARAMETRELER
# =========================
BATCH = 16
IMG_SIZE = (224, 224)
AUTOTUNE = tf.data.AUTOTUNE

DATASET_PATH = '/content/drive/MyDrive/FaceForensics_Data/Hoca_Image_Dengeli_V7'

# =========================
# DATASET LOAD
# =========================
train_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "train"),
    label_mode="binary",
    class_names=['fake', 'real'],
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=True,
    seed=123
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATASET_PATH, "val"),
    class_names=['fake', 'real'],
    label_mode="binary",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=True
)

train_ds = train_ds.cache().shuffle(3000).prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)

# =========================
# AUGMENTATION (ÖNEMLİ)
# =========================
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.GaussianNoise(0.01),   # 🔥 dataset farkını azaltır
])

# =========================
# MODEL (EfficientNetB0)
# =========================
base = EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3)
)

# 🔥 Fine-tuning
base.trainable = True

for layer in base.layers[:-30]:
    layer.trainable = False

# =========================
# MODEL HEAD
# =========================
inputs = tf.keras.Input(shape=(224, 224, 3))

x = data_augmentation(inputs)
x = layers.Lambda(preprocess_input)(x)

x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)

x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.3)(x)

outputs = layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)

# =========================
# COMPILE
# =========================
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),  # 🔥 düşük LR
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.AUC(name="auc")
    ]
)

model.summary()

# =========================
# CALLBACKS
# =========================
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=8,
        restore_best_weights=True,
        verbose=1
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.3,
        patience=3,
        min_lr=1e-7,
        verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/FaceForensics_Data/best_model_final.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

# =========================
# TRAIN
# =========================
print("🚀 MODEL BAŞLIYOR")

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=40,
    callbacks=callbacks
)
```
